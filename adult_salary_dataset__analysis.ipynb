```python
[q1,q3] = df['capital-gain'].describe().loc[['25%','75%']]
IQR = q3-q1
lb = q1-(1.5*IQR)
ub = q3 + (1.5*IQR)
print('IQR',IQR)
print('lower boundary',lb)
print('upper boundary',ub)
print('any outliers',df['capital-gain'][(df['capital-gain']<lb)&(df['capital-gain']>ub)].any())
```
