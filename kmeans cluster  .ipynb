```python
from sklearn.datasets import make_blobs
```

```python
datapoints = make_blobs(n_samples=200,n_features=2,centers=8,cluster_std=2,random_state=7)
```

```python
datapoints[0]
```

```python
datapoints[1]
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(datapoints[0][:,0],datapoints[0][:,1],c=datapoints[1])
```

```python
from sklearn.cluster import KMeans
```

```python
kmodel = KMeans(n_clusters=8)
```

```python
kmodel.fit(datapoints[0])
```

```python
kmodel.cluster_centers_
```

```python
kmodel.labels_
```

```python
fig, (ax1 , ax2) = plt.subplots(1,2)
ax1.set_title('original data')
ax1.scatter(datapoints[0][:,0],datapoints[0][:,1],c=datapoints[1])
ax2.set_title('kmeans cluster')
ax2.scatter(datapoints[0][:,0],datapoints[0][:,1],c=kmodel.labels_)
```
